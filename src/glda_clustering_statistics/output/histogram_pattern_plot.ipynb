```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
%matplotlib inline
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = [16, 9]
```

```python
histogram_patterns = np.loadtxt('histogram_patterns.csv', delimiter=',')
df = pd.DataFrame(histogram_patterns)
df['pattern_name'] = np.arange(1,22)
df
```

```python
df2 = df.set_index('pattern_name').T
df2
```

```python
fig = df2.plot(kind='area', subplots=True, layout=(7,3), legend=False, title='Unique histogram patterns across all channels', grid=True, figsize=(20, 10),sharey=True)
plot=fig[0][0].get_figure()
plt.tight_layout()
plot.savefig('unique_histograms.png')
```
