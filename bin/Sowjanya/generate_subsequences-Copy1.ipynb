# Finding the statistics of the subsequences(mean,variance,skewness,IQR)


```python
def subsequence_statistics(n):
    subsequences=np_sequences[n]
    Mean=[]
    Standard_deviation=[]
    Skewness=[]
    IQR=[]
    Min=[]
    Max=[]
    Median=[]
    Range=[]
    Lower_quartile=[]
    Middle_quartile=[]
    Upper_quartile=[]
    Coefficient_of_variation=[]
    Kurtosis=[]
    for i in range(0,len(subsequences)):
        
        mean=sum(subsequences[i])/len(subsequences[i])
        Mean.append(mean)
        
        std=statistics.stdev(subsequences[i])
        Standard_deviation.append(std)
        
        Cov=std/mean
        Coefficient_of_variation.append(Cov)
        
        minimum=min(subsequences[i])
        Min.append(minimum)
        
        maximum=max(subsequences[i])
        Max.append(maximum)
        
        range1=maximum-minimum
        Range.append(range1)
        
        skewness=stats.skew(subsequences[i])
        Skewness.append(skewness)
        
        median=statistics.median(subsequences[i])
        Median.append(median)
        
        q3,q2, q1 = np.percentile(subsequences[i], [75 ,50,25])
        
        Lower_quartile.append(q1)
        
        Middle_quartile.append(q2)
        
        Upper_quartile.append(q3)
        
        iqr = q3 - q1
        IQR.append(iqr)
        
        kurtosis=stats.kurtosis(subsequences[i])
        Kurtosis.append(kurtosis)
        
        
        
    return Mean,Variance,Skewness,IQR,Min,Max,Median,Range,Lower_quartile,Middle_quartile,Upper_quartile,Coefficient_of_variation,Kurtosis
    
```

```python
Mean,Variance,Skewness,IQR,Min,Max,Median,Range,Lower_quartile,Middle_quartile,Upper_quartile,Coefficient_of_variation,Kurtosis=subsequence_statistics(0)
data = list(zip(Mean,Variance,Skewness,IQR,Min,Max,Median,Range,Lower_quartile,Middle_quartile,Upper_quartile,Coefficient_of_variation,Kurtosis))
X1_statistic_df=pd.DataFrame(data,columns=['Mean','Variance','Skewness','IQR','Min','Max','Median','Range','Lower_quartile','Middle_quartile','Upper_quartile','Coefficient_of_variation','Kurtosis'])
```
